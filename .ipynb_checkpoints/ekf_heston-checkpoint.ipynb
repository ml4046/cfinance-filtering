```python
def std_normal_sample():
    u1 = np.random.rand()
    u2 = np.random.rand()
    v1 = 2*u1 - 1
    v2 = 2*u2 - 1
    s = v1**2 + v2**2
    x1 = np.sqrt(-2*np.log(s)/s) * v1
    x2 = np.sqrt(-2*np.log(s)/s) * v2
    while s >= 1:
        u1 = np.random.rand()
        u2 = np.random.rand()
        v1 = 2*u1 - 1
        v2 = 2*u2 - 1
        s = v1**2 + v2**2
        x1 = np.sqrt(-2*np.log(s)/s) * v1
        x1 = np.sqrt(-2*np.log(s)/s) * v2
    return x1, x2
```
